```python
#Task-03
#Password Complexity Checker
'''Build a tool that assesses the strength of a password based on criteria 
such as length, presence of uppercase and lowercase letters, numbers,
and special characters. Provide feedback to users on the password's strength.'''


#tkinter is used to construct basic ghraphical user interface(GUI)
#re module is used for regular expression in python
import tkinter as tk
from tkinter import messagebox
import re

#function to check the strength of password based on some criteria
def assess_password_strength(password):
    criteria = {
        "length": len(password) >= 8,
        "uppercase": any(c.isupper() for c in password),
        "lowercase": any(c.islower() for c in password),
        "digit": any(c.isdigit() for c in password),
        "special": any(c in "!@#$%^&*()-_=+[]{}|;:,.<>?/" for c in password)
    }

    strength = sum(criteria.values())

    if strength == 5:
        feedback = "Very Strong"
    elif strength == 4:
        feedback = "Strong"
    elif strength == 3:
        feedback = "Medium"
    elif strength == 2:
        feedback = "Weak"
    else:
        feedback = "Very Weak"

    suggestions = []
    if not criteria["length"]:
        suggestions.append("Password should be at least 8 characters long.")
    if not criteria["uppercase"]:
        suggestions.append("Include at least one uppercase letter.")
    if not criteria["lowercase"]:
        suggestions.append("Include at least one lowercase letter.")
    if not criteria["digit"]:
        suggestions.append("Include at least one digit.")
    if not criteria["special"]:
        suggestions.append("Include at least one special character (!@#$%^&*()-_=+[]{}|;:,.<>?/).")

    return feedback, suggestions

def check_password_strength():
    password = entry.get()
    feedback, suggestions = assess_password_strength(password)
    messagebox.showinfo("Password Strength", f"Your password strength is: {feedback}\n\nSuggestions:\n" + "\n".join(suggestions))

# Setting up the main window
root = tk.Tk()
root.title("Password Strength Checker")
root.geometry("400x200")

# Adding a label
label = tk.Label(root, text="Enter your password:", font=("Helvetica", 14))
label.pack(pady=10)

# Adding an entry field
entry = tk.Entry(root, show="", width=30, font=("Helvetica", 12))
entry.pack(pady=10)

# Adding a button
button = tk.Button(root, text="Check Strength", command=check_password_strength, font=("Helvetica", 12))
button.pack(pady=20)

# Running the GUI application
root.mainloop()

```
